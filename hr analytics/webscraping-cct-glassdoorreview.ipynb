```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract(page):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36'}
    url = f'https://www.glassdoor.ca/Reviews/Critical-Control-Reviews-E23461_P{i}.htm?filter.iso3Language=eng'
    r = requests.get(url, headers=headers)
    content = r.content
    soup = BeautifulSoup(content, 'html.parser')
    return soup
def transform(soup):
    divs = soup.find_all('div', attrs={'class':'gdReview'})
    for item in divs:
        title = item.find('a', attrs={'class': 'reviewLink'}).text
        rating = item.find('span', attrs={'class': 'ratingNumber mr-xsm'}).text
        details = item.find('span', attrs={'class': 'authorJobTitle middle common__EiReviewDetailsStyle__newGrey'}).text
        location = item.find('span', attrs={'class': 'authorLocation'})
        current = item.find('span', attrs={'class': 'pt-xsm pt-md-0 css-1qxtz39 eg4psks0'}).text
        pros = item.find('span', attrs={'data-test': 'pros' }).text
        cons = item.find('span', attrs={'data-test': 'cons' }).text
        review = {
            'Title':title,
            'Rating':rating,
            'Details': details,
            'Location': location,
            'Pros':pros,
            'Cons':cons,
            'Current': current
        }
        reviewlist.append(review)
    return

reviewlist = []

for i in range(0,10,1):
    print(f'Getting page, {i}')
    c = extract(0)
    transform(c)

df = pd.DataFrame(reviewlist)

print(df.head())
df.to_csv('review-glassdoor-cct.csv', index = False)
```
