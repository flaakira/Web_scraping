```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
import datetime as dt
```

# Period Analyses

```python
start = dt.datetime(2018,1,1)
end = dt.datetime.today()
```

## Cervus Stock

```python
stock_cerv = web.DataReader('cerv.to','yahoo', start, end).reset_index()
```

```python
stock_cerv.tail()
```

```python
plt.figure(figsize=(18,8))
plt.plot(stock_cerv['Date'], stock_cerv['Open'], color = 'r')
plt.plot(stock_cerv['Date'], stock_cerv['Close'], color = 'b')
plt.plot(stock_cerv['Date'], stock_cerv['High'], color = 'y')
plt.plot(stock_cerv['Date'], stock_cerv['Low'], color = 'g')
plt.title('CERV.TO')
plt.show()
```
